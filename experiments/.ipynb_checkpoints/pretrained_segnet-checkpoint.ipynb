```python
class SegmDataset(torch.utils.data.Dataset):
    def __init__(self, root, tr):
        self.root = root
        self.transforms = tr
        # load all image files, sorting them to
        # ensure that they are aligned
        self.imgs = list(sorted(os.listdir(os.path.join(root, "Images_1"))))
        self.masks = list(sorted(os.listdir(os.path.join(root, "Masks_1"))))

    def __getitem__(self, idx):
        # load images and masks
        img_path = os.path.join(self.root, "Images_1", self.imgs[idx])
        mask_path = os.path.join(self.root, "Masks_1", self.masks[idx])
        img = Image.open(img_path).convert('RGB')
        mask = Image.open(mask_path).convert('L')

        image_id = idx

        img, target = self.transforms(img), self.transforms(mask)

        return img, torch.round(torch.sigmoid(target))

    def __len__(self):
        return len(self.imgs)
```

```python
dataset = SegmDataset(dataDir, transformer)
```

```python
train_size = int(TRAIN_SIZE * len(dataset))
valid_size = len(dataset) - train_size

train_dataset, valid_dataset = torch.utils.data.random_split(dataset, [train_size, valid_size], generator=generator)

train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=0)
valid_dataloader = torch.utils.data.DataLoader(valid_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=0)
```

## SegNet Architecture

```python
class SegNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.enc_conv0 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64)
        )
        self.pool0 = nn.MaxPool2d(2, stride=2, return_indices=True) # 256 -> 128

        self.enc_conv1 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128)
        )

        self.pool1 = nn.MaxPool2d(2, stride=2, return_indices=True) # 128 -> 64

        self.enc_conv2 = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256)
        )

        self.pool2 = nn.MaxPool2d(2, stride=2, return_indices=True) # 64 -> 32
        self.enc_conv3 = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512)
        )
        self.pool3 = nn.MaxPool2d(2, stride=2, return_indices=True) # 32 -> 16

        # bottleneck
        self.bottleneck_conv = nn.Sequential(
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512)
        )

        # decoder (upsampling)
        self.upsample0 = nn.MaxUnpool2d(2, stride=2) # 16 -> 32
        self.dec_conv0 = nn.Sequential(
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.Conv2d(512, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256)
        )
        self.upsample1 = nn.MaxUnpool2d(2, stride=2) # 32 -> 64
        self.dec_conv1 = nn.Sequential(
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.Conv2d(256, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128)
        )
        self.upsample2 = nn.MaxUnpool2d(2, stride=2)  # 64 -> 128
        self.dec_conv2 = nn.Sequential(
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.Conv2d(128, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64)
        )
        self.upsample3 = nn.MaxUnpool2d(2, stride=2)  # 128 -> 256
        self.dec_conv3 = nn.Sequential(
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.Conv2d(64, 1, kernel_size=3, padding=1),
            nn.BatchNorm2d(1)
        )

    def forward(self, x):
        # encoder
        e0 = F.relu(self.enc_conv0(x))
        e0, ind0 = self.pool0(e0)
        e1 = F.relu(self.enc_conv1(e0))
        e1, ind1 = self.pool1(e1)
        e2 = F.relu(self.enc_conv2(e1))
        e2, ind2 = self.pool2(e2)
        e3 = F.relu(self.enc_conv3(e2))
        e3, ind3 = self.pool3(e3)

        # bottleneck
        b = F.relu(self.bottleneck_conv(e3))

        # decoder
        d0 = self.upsample0(b, ind3)
        d0 = F.relu(self.dec_conv0(d0))
        d1 = self.upsample1(d0, ind2)
        d1 = F.relu(self.dec_conv1(d1))
        d2 = self.upsample2(d1, ind1)
        d2 = F.relu(self.dec_conv2(d2))
        d3 = self.upsample2(d2, ind0)
        d3 = self.dec_conv3(d3) # no activation
        return d3
```

## Unet Architecture

```python
class UNet(nn.Module):
    def __init__(self):
        super().__init__()

       
        self.enc_conv0 = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 1),
            nn.BatchNorm2d(64),
            nn.ReLU()
        )
        self.pool0 = nn.MaxPool2d(3, 2, padding=1)
        self.enc_conv1 = nn.Sequential(
            nn.Conv2d(64, 128, 3, 1, 1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, 3, 1, 1),
            nn.BatchNorm2d(128),
            nn.ReLU()
        )
        self.pool1 = nn.MaxPool2d(3, 2, padding=1)
        self.enc_conv2 = nn.Sequential(
            nn.Conv2d(128, 256, 3, 1, 1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Conv2d(256, 256, 3, 1, 1),
            nn.BatchNorm2d(256),
            nn.ReLU()
        )
        self.pool2 = nn.MaxPool2d(3, 2, padding=1)
        self.enc_conv3 = nn.Sequential(
            nn.Conv2d(256, 512, 3, 1, 1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.Conv2d(512, 512, 3, 1, 1),
            nn.BatchNorm2d(512),
            nn.ReLU()
        )
        self.pool3 = nn.MaxPool2d(3, 2, padding=1)

        # bottleneck
        self.bottleneck_conv = nn.Sequential(
            nn.Conv2d(512, 1024, 3, 1, 1),
            nn.ReLU(),
            nn.BatchNorm2d(1024),
            nn.Conv2d(1024, 512, 3, 1, 1),
            nn.BatchNorm2d(512),
            nn.ReLU()
        )

        # decoder (upsampling)
        self.upsample0 = nn.Upsample(scale_factor=2, mode='nearest')
        self.dec_conv0 = nn.Sequential(
            nn.Conv2d(1024, 512, 3, 1, 1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.Conv2d(512, 256, 3, 1, 1),
            nn.BatchNorm2d(256),
            nn.ReLU()
        )
        self.upsample1 = nn.Upsample(scale_factor=2, mode='nearest')
        self.dec_conv1 = nn.Sequential(
            nn.Conv2d(512, 256, 3, 1, 1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Conv2d(256, 128, 3, 1, 1),
            nn.BatchNorm2d(128),
            nn.ReLU()
        )
        self.upsample2 = nn.Upsample(scale_factor=2, mode='nearest')
        self.dec_conv2 = nn.Sequential(
            nn.Conv2d(256, 128, 3, 1, 1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 64, 3, 1, 1),
            nn.BatchNorm2d(64),
            nn.ReLU()
        )
        self.upsample3 = nn.Upsample(scale_factor=2, mode='nearest')
        self.dec_conv3 = nn.Sequential(
            nn.Conv2d(128, 64, 3, 1, 1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 1, 1)
        )

    def forward(self, x):
        # encoder
        e0 = self.enc_conv0(x)
        e1 = self.enc_conv1(self.pool0(e0))
        e2 = self.enc_conv2(self.pool1(e1))
        e3 = self.enc_conv3(self.pool2(e2))
        # bottleneck
        b = self.bottleneck_conv(self.pool3(e3))

        # decoder
        x = self.dec_conv0(torch.cat([self.upsample0(b), e3], dim=1))
        x = self.dec_conv1(torch.cat([self.upsample1(x), e2], dim=1))
        x = self.dec_conv2(torch.cat([self.upsample2(x), e1], dim=1))
        x = self.dec_conv3(torch.cat([self.upsample3(x), e0], dim=1))  # no activation
        return x
```

## Work Functions

```python
def iou_pytorch(outputs: torch.Tensor, labels: torch.Tensor):
    outputs = torch.round(torch.sigmoid(outputs))
    outputs = outputs.squeeze(1).byte()  # BATCH x 1 x H x W => BATCH x H x W
    labels = labels.squeeze(1).byte()
    SMOOTH = 1e-8
    intersection = (outputs & labels).float().sum((1, 2))  # Will be zero if Truth=0 or Prediction=0
    union = (outputs | labels).float().sum((1, 2))         # Will be zzero if both are 0

    iou = (intersection + SMOOTH) / (union + SMOOTH)  # We smooth our devision to avoid 0/0

    thresholded = torch.clamp(20 * (iou - 0.5), 0, 10).ceil() / 10  # This is equal to comparing with thresolds

    return thresholded  #
```

```python
def bce_loss(y_real, y_pred):
    result = y_pred - y_real * y_pred + torch.log(1 + torch.exp(-y_pred))
    return result.mean()
```

```python
def train(model, opt, loss_fn, epochs, data_tr, data_val):
    graph_loss = []
    graph_loss_val = []
    graph_score = []
    val_set = next(iter(data_val))
    for epoch in range(epochs):
        tic = time()
        print('* Epoch %d/%d' % (epoch+1, epochs))

        avg_loss = 0
        model.train()  # train mode
        for X_batch, Y_batch in data_tr:
            X_batch = X_batch.to(device)
            Y_batch = Y_batch.to(device)
            opt.zero_grad()
            # forward
            Y_pred = model(X_batch)
            loss =  loss_fn(Y_pred, Y_batch)
            loss.backward()
            opt.step()

            # calculate loss to show the user
            avg_loss += loss / len(data_tr)
        toc = time()
        print('train loss: %f' % avg_loss)
        graph_loss.append(avg_loss.cpu().detach().numpy())

        # show intermediate results
        model.eval()  # testing mode
        avg_loss_val = 0
        X_val, Y_val = val_set
        X_val = X_val.to(device)
        Y_val = Y_val.to(device)
        with torch.no_grad():
          pred = model(X_val)
          pred = torch.round(torch.sigmoid(pred))
          Y_hat = pred.cpu().detach().numpy()
          loss_val =  loss_fn(pred, Y_val)
          avg_loss_val += loss_val / len(val_set)
        print('validation loss: %f' % avg_loss_val)
        graph_loss_val.append(avg_loss_val.cpu().detach().numpy())
        score = iou_pytorch(pred, Y_val).mean().item() / len(val_set)
        graph_score.append(score)
        print('validation IoU: %f' % score)

        # Visualize tools
        clear_output(wait=True)
        for k in range(4):
            plt.subplot(2, 6, k+1)
            plt.imshow(np.rollaxis(X_val[k].cpu().numpy(), 0, 3), cmap='gray')
            plt.title('Real')
            plt.axis('off')

            plt.subplot(2, 6, k+7)
            plt.imshow(Y_hat[k, 0], cmap='gray')
            plt.title('Output')
            plt.axis('off')
        plt.suptitle('%d / %d - loss: %f' % (epoch+1, epochs, avg_loss))
        plt.show()
    return graph_loss, graph_loss_val, graph_score
```

## SegNet

```python
opt = torch.optim.Adam(segnet.parameters(), lr=1e-4)
criterion = bce_loss
epoch = 50
result, result_val, score = train(segnet, opt, criterion, epoch, train_dataloader, valid_dataloader)
plt.plot(result)
plt.plot(result_val)
plt.legend(['train', 'val'])
plt.grid()
plt.title(f'{epoch} EPOCHS')
plt.ylabel('BCE loss')
plt.xlabel('number of epochs')
name = f'segnet_{epoch}.png'
plt.savefig(dataDir + name)
plt.show()
plt.plot(score)
plt.grid()
plt.ylabel('IoU')
plt.xlabel('number of epochs')
plt.show()
```

```python
images, targets = next(iter(train_dataloader))
images = images.to(device)
segnet.eval()
predictions = segnet(images)
```

```python
plt.figure(figsize=(18, 20))
for i in range(8):
    plt.subplot(4, 2, i+1)
    plt.axis("off")
    image = images[i]
    mask = predictions[i]
    mask = torch.round(torch.sigmoid(mask)).to(torch.bool)
    image = (255.0 * (image - image.min()) / (image.max() - image.min())).to(torch.uint8)
    output_image = draw_segmentation_masks(image, mask, alpha=0.5, colors="blue").cpu().detach()
    plt.imshow(output_image.permute(1,2,0))
plt.show();
```
